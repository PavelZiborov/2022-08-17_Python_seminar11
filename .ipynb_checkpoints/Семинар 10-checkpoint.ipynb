```python
# 5x^2 + 10x - 30
# Дана функция f(x) = 5x^2 + 10x - 30 

# 1. Определить корни

# 2. Найти интервалы, на которых функция возрастает

# 3. Найти интервалы, на которых функция убывает

# 4. Построить график

# 5. Вычислить вершину

# 6. Определить промежутки, на котором f > 0 (производная)

# 7. Определить промежутки, на котором f < 0 (производная)
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl

x = list(range(-5, 5))
fx = [5*i**2 + 10*i - 30 for i in x]
# fx = [-12*i**4 * sin(cos(i)) - 18*i**3 + 5*i**2 +10*i -30  for i in x]

# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.html?highlight=pyplot#
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.axes.html#matplotlib.pyplot.axes

mpl.rcParams['lines.linewidth'] = 1
mpl.rcParams['lines.linestyle'] = '-' #'--'
mpl.rcParams['lines.color'] = 'C0'

# mpl.axhline(y=0, color='k')
# mpl.axvline(y=0, color='k')


plt.axes(facecolor='white')
plt.plot(x,fx)
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
a = 5
b = 10
c = -30
 
x = np.linspace(-50, 50)
y = a*x**2 + b*x + c  
 
fig, ax = plt.subplots()
ax.plot(x, y)
plt.grid(True)
plt.plot(-1, -35, 'bo')
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import sympy

x = list(range(-200, 200))

fx = [-12*i**4 * sympy.sin(sympy.cos(i)) - 18*i**3 + 5*i**2 +10*i -30  for i in x]



mpl.rcParams['lines.linewidth'] = 1
mpl.rcParams['lines.linestyle'] = '-' #'--'
mpl.rcParams['lines.color'] = 'C0'



plt.axes(facecolor='white')
plt.plot(x,fx)
plt.grid()
plt.show()
```
